```python
# Managing outliers
# Option 1: Remove outliers using IQR filtering
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

# Option 2: Cap outliers to upper/lower bounds
def cap_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = data[column].clip(lower=lower_bound, upper=upper_bound)
    return data

# Choose method: Remove or Cap
for col in numerical_cols:
    df = cap_outliers(df, col)  # Use this to cap outliers
    # df = remove_outliers_iqr(df, col)  # Uncomment this to remove outliers

print("\nDataset after managing outliers:")
print(df.describe())
```
